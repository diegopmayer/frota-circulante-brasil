```python
import pandas as pd
%matplotlib inline
```

```python
with pd.ExcelFile('Frota_Reg_UF-Tipo_Modelo_Fevereiro_2019.xls') as xls:
    df_estados = pd.read_excel(xls, usecols=[0, 4, 8, 15, 16], na_values=['NA'], skiprows=2, skipfooter=10)
    df_regioes = pd.read_excel(xls, usecols=[0, 4, 8, 15, 16], na_values=['NA'], skiprows=2, skipfooter=10) 
```

```python
# Renomear Colunas
df_estados.columns = ['Estado', 'Caminhões', 'Plataforma', 'Reboque', 'Semi-Reboque']
df_regioes.columns = ['Região', 'Caminhões', 'Plataforma', 'Reboque', 'Semi-Reboque']   
```

```python
# Criar campo de soma Total
df_estados['Total'] = (df_estados['Caminhões'] + df_estados['Reboque'] + df_estados['Semi-Reboque'] + df_estados['Plataforma'])
df_regioes['Total'] = (df_regioes['Caminhões'] + df_regioes['Reboque'] + df_regioes['Semi-Reboque'] + df_regioes['Plataforma'])
```
